```python
import pandas as pd
import numpy as np

df = pd.read_csv('laliga.csv',index_col=False, skiprows=1)

def Goal_diff_count():
    print("Goal Difference")    
    df2=df
    df2['GD']=df2.apply(lambda x:x['GoalsFor']-x['GoalsAgainst'],axis=1)
    print(df2)
    df2=df2.nlargest(len(df2), 'GD')
    #print(df2)
    print("Maximum Goal Difference: ", df2.iloc[0,1])
    print("Minimum Goal Difference: ", df2.iloc[len(df2)-1,1])
    

#print(df)
#print(len(df))

df=df.replace('-',0)
print ("After replacing '-' with 0")
print(df.columns)
df['Debut'] = df['Debut'].apply(lambda x: x[0:4])

df['Debut'] = df['Debut'].astype('int64')

df['GoalsFor'] = df['GoalsFor'].astype('int64')

df['GoalsAgainst'] = df['GoalsAgainst'].astype('int64')

df['GamesWon'] = df['GamesWon'].astype('int64')

df['GamesPlayed'] = df['GamesPlayed'].astype('int64')

df['Debut'] = df['Debut'].astype(np.int64)

print(df.iloc[0,16])
print('Teams Debut between 1930 and 1980')

for i in range(0,len(df)):
    if df.iloc[i,17]>=1930 and df.iloc[i,17]<=1980:
        print(df.iloc[i,1]) 


df['Points'] = df['Points'].astype('int64')


df1=df.nlargest(5, 'Points')
count=0
print("Top 5 Teams on point basis")
for v in df['Team']:
    count=count+1
    print(v)
    if count==int(5):
        break;

Goal_diff_count()


df['Winning Percent']=df.apply(lambda x:(x['GamesWon']/x['GamesPlayed'])*100 if x['GamesPlayed']!=0 else 0,axis=1)

df3=df.nlargest(len(df), 'Winning Percent')

print("Top 5 Teams on goal difference basis")

print(df3['Team'].head(5))

#print("Best")
r = df.groupby('BestPosition')['Points'].sum()
print("Sum of points based on 'Best Position")
print (r)


#df3=df.groupby('BestPosition')

#for pos,

df.to_csv('output.csv',index=False)
```
